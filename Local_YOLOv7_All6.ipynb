# Convert labels to sumission format

```python
img_size = {}
for c in test_countries:
    files = glob.glob(os.path.join(input_dir, c, "test/images", "*.jpg"))
    im = cv2.imread(files[0])
    img_size[c] = im.shape
```

```python
def convert_yolo_to_pascal(yolo, img_size, c):
    img_h, img_w = img_size[c][0], img_size[c][1]
    c_x, c_y, w, h = yolo
    c_x1 = c_x*img_w
    c_y1 = c_y*img_h
    w1 = w*img_w
    h1 = h*img_h
    x_min = int(c_x1 - w1/2)
    y_min = int(c_y1 - h1/2)
    x_max = int(c_x1 + w1/2)
    y_max = int(c_y1 + h1/2)
    
    if x_min < 0:
        x_min = 0
    if y_min < 0:
        y_min = 0
    if x_max > img_w:
        x_max = img_w
    if y_max > img_h:
        y_max = img_h
    
    return [x_min, y_min, x_max, y_max]
```

```python
import pdb
def convert_to_submission(file, img_size, c):
    img_id = os.path.splitext(os.path.basename(file))[0] + ".jpg"
    o_str = ""
    with open(file, "r") as f:
        output_str = []
        lines = f.read().splitlines()
        for line in lines:
            parts = line.split(' ')
            class_lbl = int(parts[0]) + 1
            yolo = [float(_) for _ in parts[1:]]
            pascal = convert_yolo_to_pascal(yolo, img_size, c)
            output_str += ([class_lbl] + pascal)
        o_str = " ".join([str(_) for _ in output_str])
    return img_id,o_str
```

```python
convert_to_submission("runs/detect/model1_China_MotorBike/labels/China_MotorBike_001978.txt", img_size, "China_MotorBike")
```

```python
def get_results_for_countries(countries, model_name, output_file):
    all_detected = {}
    for c in countries:
        files = glob.glob(f"runs/detect/{model_name}_{c}/labels/*.txt")
        for f in files:
            img_id, o_str = convert_to_submission(f, img_size, c)
            all_detected[img_id] = o_str
    all_outs = []
    
    for c in countries:
        files = glob.glob(os.path.join(input_dir, c, "test/images", "*.jpg"))
        for f in files:
            img_id = os.path.basename(f)
            o_str = all_detected[img_id] if img_id in all_detected else ""
            all_outs.append(f"{img_id},{o_str}")
    
    with open(output_file, "w") as f:
        for l in all_outs:
            f.write(l + "\n")
        print(f"Wrote {output_file}")
    
```

```python
for c in test_countries:
    get_results_for_countries([c], model_name, c+".txt")
```
