```python
def convert_yolo_to_pascal(yolo, img_h, img_w, img_h_o, img_w_o, c):
    c_x, c_y, w, h = yolo
    c_x1 = c_x*img_w
    c_y1 = c_y*img_h
    w1 = w*img_w
    h1 = h*img_h
    x_min = int(c_x1 - w1/2)
    y_min = int(c_y1 - h1/2)
    x_max = int(c_x1 + w1/2)
    y_max = int(c_y1 + h1/2)
    
    if x_min < 0:
        x_min = 0
    if y_min < 0:
        y_min = 0
    if x_max > img_w:
        x_max = img_w
    if y_max > img_h:
        y_max = img_h
    if c == "Norway1":
        y_min = y_min + (img_h_o-1824)
        y_max = y_max + (img_h_o-1824)
    return [x_min, y_min, x_max, y_max]
```

```python
from PIL import Image
def convert_to_submission(file, c, min_conf={1: 0, 2:0, 3:0, 4:0}):
    img_id = os.path.splitext(os.path.basename(file))[0] + ".jpg"
    o_str = ""
    with open(file, "r") as f:
        output_str = []
        lines = f.read().splitlines()
        outputs = []
        img = Image.open(os.path.join(input_dir,c, "test/images", img_id))
        img_w, img_h = img.size
        img_w_o = img_w
        img_h_o = img_h
        if c == "Norway1":
            img_w_o, img_h_o = Image.open(os.path.join(input_dir, "Norway", "test/images", img_id)).size
            
        for line in lines:
            parts = line.split(' ')
            class_lbl = int(parts[0]) + 1
            yolo = [float(_) for _ in parts[1:5]]
            conf = 0
            if len(parts)==6:
                conf = float(parts[5])
            if conf >= min_conf[class_lbl]: # only add results if it is greater than min conf
              # # switch label
              # if class_lbl == 1 and conf <= min_conf[class_lbl] + 0.1:
              #   class_lbl = 2
              outputs.append([class_lbl, yolo, conf])

            # else:
            #   if conf >= 0.375 and class_lbl == 1:
            #       class_lbl = 2
            #       outputs.append([class_lbl, yolo, conf])
            #   if conf >= 0.32 and class_lbl == 2:
            #     class_lbl = 1
            #     outputs.append([class_lbl, yolo, conf])
                

        
        outputs.sort(key=lambda x: -x[2]) # sort by higher confidence first
        for x in outputs:
            class_lbl = x[0]
            yolo = x[1]
            pascal = convert_yolo_to_pascal(yolo, img_h, img_w, img_h_o, img_w_o, c)
            output_str += ([class_lbl] + pascal)
        o_str = " ".join([str(_) for _ in output_str])
    return img_id,o_str
```

```python
import pdb
def get_results_for_countries(countries, detect_model_name, output_file, runs="", min_conf={1:0,2:0, 3:0,4:0}):
    all_detected = {}
    for c in countries:
        files = glob.glob(f"runs/detect/{detect_model_name}{runs}/labels/*.txt")
        for f in files:
            img_id, o_str = convert_to_submission(f, c, min_conf)
            all_detected[img_id] = o_str
    all_outs = []
    
    for c in countries:
        files = glob.glob(os.path.join(input_dir, c, "test/images", "*.jpg"))
        for f in files:
            img_id = os.path.basename(f)
            o_str = all_detected[img_id] if img_id in all_detected else ""
            all_outs.append(f"{img_id},{o_str}")
    
    
    with open(output_file, "w") as f:
        for l in all_outs:
            f.write(l + "\n")
        print(f"Wrote {output_file}")
    
```

```python
detect_model_name = "United_States"
detect_runs = "14" # best Japan + Us + India
```

```python
for x in (0.45, 0.50, 0.55): #(.505, .515): #(0.48, 0.49, 0.51, 0.52): #(0.25, 0.30, 0.35,0.40, 0.45, 0.50, 0.55):# (0.48, 0.49, 0.51, 0.52, 0.60): #(0.35,0.40, 0.45, 0.50, 0.55):
    get_results_for_countries([c], detect_model_name, c+"{:.3f}".format(x)+".txt", runs=detect_runs, min_conf={1: x, 2:x, 3:x, 4:x}) # best so far at 0.50 (runs5)
```
